```python
import boto3
import pandas as pd

s3_client = boto3.client('s3')

s3_resource = boto3.resource(
    service_name='s3',
    region_name='ap-south-1')
```

## read file in s3 bucket

```python
table_one = s3_resource.Bucket('vipul-first-bucket').Object('covid data/enigma-jhud/csv/Enigma-JHU.csv').get()
table_two = s3_resource.Bucket('vipul-first-bucket').Object('covid data/enigma-nytimes-data-in-usa/csv/us_county/us_county.csv.fDde7E2b').get()
table_three = s3_resource.Bucket('vipul-first-bucket').Object('covid data/enigma-nytimes-data-in-usa/csv/us_states/us_states.csv').get()
table_four = s3_resource.Bucket('vipul-first-bucket').Object('covid data/rearc-covid-19-testing-data/csv/states_daily/states_daily.csv.DD2BE07f').get()
table_five = s3_resource.Bucket('vipul-first-bucket').Object('covid data/rearc-covid-19-testing-data/csv/us_daily/us_daily.csv').get()
table_six = s3_resource.Bucket('vipul-first-bucket').Object('covid data/rearc-covid-19-testing-data/csv/us-total-latest/us.csv').get()
table_seven = s3_resource.Bucket('vipul-first-bucket').Object('covid data/static-datasets/csv/countrycode/CountryCodeQS.csv').get()
table_eight = s3_resource.Bucket('vipul-first-bucket').Object('covid data/static-datasets/csv/CountyPopulation/County_Population.csv').get()
table_nine = s3_resource.Bucket('vipul-first-bucket').Object('covid data/static-datasets/csv/state-abv/states_abv.csv').get()
## json file
table_ten = s3_resource.Bucket('vipul-first-bucket').Object('covid data/rearc-usa-hospital-beds/usa-hospital-beds.json').get()
```

```python
Enigma_JHU = pd.read_csv(table_one['Body'])
us_county = pd.read_csv(table_two['Body'])
us_states = pd.read_csv(table_three['Body'])
states_daily = pd.read_csv(table_four['Body'])
us_daily = pd.read_csv(table_five['Body'])
us = pd.read_csv(table_six['Body'])
country_code = pd.read_csv(table_seven['Body'])
County_Population = pd.read_csv(table_eight['Body'])
states_abv = pd.read_csv(table_nine['Body'])
usa_hospital_beds = pd.read_json(table_ten['Body'], lines=True)
```

## build starschema 

## fact_covid

```python
Enigma_JHU.head()
```

```python
states_daily.head()
```

```python
states_daily.columns
```

```python
fact_covid_one = Enigma_JHU[['fips', 'province_state', 'country_region', 'confirmed', 'deaths', 'recovered', 'active']] 
fact_covid_one.head()
```

```python
fact_covid_two = states_daily[['fips', 'date', 'positive', 'negative', 'hospitalized', 'hospitalizedDischarged', 'hospitalizedCurrently']]
fact_covid_two.head()
```

```python
fact_covid = pd.merge(fact_covid_one, fact_covid_two, on='fips', how='inner')
fact_covid.head()
```

```python
fact_covid.tail()
```

```python
fact_covid.to_csv("fact_covid.csv", index=False)
```

## dim_region

```python
Enigma_JHU.head()
```

```python
us_county.head()
```

```python
dim_region_one = Enigma_JHU[['fips', 'province_state', 'country_region', 'latitude', 'longitude']] 
dim_region_one.head()
```

```python
dim_region_two = us_county[['fips', 'county', 'state']]
dim_region_two.head()
```

```python
dim_region = pd.merge(dim_region_one, dim_region_two, on='fips', how='inner')
dim_region.head()
```

```python
dim_region.to_csv("dim_region.csv", index=False)
```

## dim_hospital

```python
usa_hospital_beds.columns
```

```python
dim_hospital = usa_hospital_beds[['FIPS', 'STATE_NAME', 'latitude', 'longtitude', 'HOSPITAL_TYPE', 'HOSPITAL_NAME', 'HQ_CITY', 'HQ_STATE']]
dim_hospital.head()
```

```python
dim_hospital.to_csv("dim_hospital.csv", index=False)
```

## dim_date

```python
states_daily.columns
```

```python
dim_date = states_daily[['fips', 'date']]
dim_date.head()
```

```python
dim_date['date'] = pd.to_datetime(dim_date['date'], format='%Y%m%d')
```

```python
dim_date.head()
```

```python
dim_date['year'] = dim_date['date'].dt.year
dim_date['month'] = dim_date['date'].dt.month
dim_date['day_of_week'] = dim_date['date'].dt.dayofweek
```

```python
dim_date.head()
```

```python
dim_date.to_csv("dim_date.csv", index=False)
```
